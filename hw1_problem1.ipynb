In the following three parts, the critical points in the 3D contour plots are designated with a red dot, so the plots are more see-through in order to see the critical points

##Initial Work

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from mpl_toolkits import mplot3d

maxFunc = 0
minFunc = 1
neitherFunc = 2
```

##Extra Functions

```
def create3DContour(X, Y, Z, pointX, pointY, pointZ, funcType):
  fig = plt.figure()
  ax = plt.axes(projection='3d')
  ax.contour3D(X, Y, Z, 50, cmap='binary', alpha=.5)
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
  
  ax.scatter(pointX, pointY, pointZ,s=50, c='r', marker='o')
  plt.show()
```

```
def create2DContour(X, Y, Z, pointX, pointY, pointZ):
  ax = plt.axes()
  plt.contour(X, Y, Z, colors='black');
  plt.scatter(pointX, pointY, pointZ,s=50, c='r', marker='o')
  plt.show()
```

##Local Min

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = X ** 2 + Y ** 2
pointX = 0
pointY = 0
pointZ = 0
create3DContour(X, Y, Z, pointX, pointY, pointZ, minFunc)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

##Local Max

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = - (X ** 2) - (Y ** 2)
pointX = 0
pointY = 0
pointZ = 0
create3DContour(X, Y, Z, pointX, pointY, pointZ, maxFunc)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

##Neither

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = (X ** 2) - (Y ** 2)
pointX = 0
pointY = 0
pointZ = 0
create3DContour(X, Y, Z, pointX, pointY, pointZ, neitherFunc)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```
